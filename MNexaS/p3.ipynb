```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import bisect, newton
```

![alt text](image.png)

```python
def f(x):
    return x**5 + 5.25*x**4 + 4.125*x**3 - 9.125*x**2 - 14.625*x -5.625 
```

```python
# grafica la función
x = np.linspace(-3, 1, 400)
y = f(x)
plt.plot(x, y, label='f(x)')
plt.axhline(0, color='red', linestyle='--', label='y=0')
plt.title('Gráfica de la función f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.show()
```

```python
def fprime(x):
    return 5*x**4 + 21*x**3 + 12.375*x**2 - 18.25*x - 14.625
```

```python
raiz1 = bisect(f, -3, 1)
print(f"Raíz encontrada con bisección: {raiz1}")
raiz2 = newton(f, -1, fprime)
print(f"Raíz encontrada con Newton-Raphson: {raiz2}")
raiz3 = bisect(f, -1, 1)
print(f"Raíz encontrada con bisección (segunda raíz): {raiz3}")
raiz4 = newton(f, 0, fprime)
print(f"Raíz encontrada con Newton-Raphson (segunda raíz): {raiz4}")
```

```python
raices_newton = []
for x0 in range(-3, 1):
    try:
        raiz = newton(f, x0, fprime, maxiter=1000, tol=1e-6)
        raiz = round(raiz, 6)  # Redondear para evitar problemas de precisión
        if raiz not in raices_newton:
            raices_newton.append(raiz)
    except RuntimeError:
        pass
print(f"Raíces encontradas con Newton-Raphson: {raices_newton}")
```
