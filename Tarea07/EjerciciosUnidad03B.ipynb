#### 1. Dados los puntos $(0,1),\;(1,5),\;(2,3)$, determine el spline cúbico.

Definimos los tramos

$S_0$ (entre $x_0=0$ y $x_1=1$) y $S_1$ (entre $x_1=1$ y $x_2=2$) 

$S_i(x)=a_i + b_i(x-x_i) + c_i(x-x_i)^2 + d_i(x-x_i)^3$.

-> $S_0(x)=a_0 + b_0 x + c_0 x^2 + d_0 x^3$

-> $S_1(x)=a_1 + b_1(x-1) + c_1(x-1)^2 + d_1(x-1)^3$

Ecuaciones:

1. $S_0(x_0)=y_0$:

$a_0 + b_0(0) + c_0(0)^2 + d_0(0)^3 = 1 \Rightarrow a_0 = 1$.

2. $S_0(x_1)=y_1$:

$a_0 + b_0(1) + c_0(1)^2 + d_0(1)^3 = 5 \Rightarrow 1 + b_0 + c_0 + d_0 = 5 \Rightarrow b_0 + c_0 + d_0 = 4$.

3. $S_1(x_1)=y_1$:

$a_1 + b_1(0) + c_1(0)^2 + d_1(0)^3 = 5 \Rightarrow a_1 = 5$.

4. $S_1(x_2)=y_2$:

$a_1 + b_1(1) + c_1(1)^2 + d_1(1)^3 = 3 \Rightarrow 5 + b_1 + c_1 + d_1 = 3 \Rightarrow b_1 + c_1 + d_1 = -2$.

5. Continuidad de la primera derivada en $x=1$:

$S_0'(x)=b_0 + 2c_0 x + 3d_0 x^2 \Rightarrow S_0'(1)=b_0 + 2c_0 + 3d_0$,

$S_1'(x)=b_1 + 2c_1(x-1) + 3d_1(x-1)^2 \Rightarrow S_1'(1)=b_1$,

 $b_0 + 2c_0 + 3d_0 = b_1$.

6. Continuidad de la segunda derivada en $x=1$:

$S_0''(x)=2c_0 + 6d_0 x \Rightarrow S_0''(1)=2c_0 + 6d_0$,

$S_1''(x)=2c_1 + 6d_1(x-1) \Rightarrow S_1''(1)=2c_1$,

 $2c_0 + 6d_0 = 2c_1$.

7. Condiciones naturales:

$S_0''(0)=2c_0 = 0 \Rightarrow c_0 = 0$.

$S_1''(2)=2c_1 + 6d_1(1) = 0 \Rightarrow 2c_1 + 6d_1 = 0$.

Resolución:

Con $c_0=0$ la ecuación (2) da $b_0 + d_0 = 4$. 

De (6) $6d_0 = 2c_1 \Rightarrow c_1 = 3d_0$.

De la condición natural en $x_2$: $2c_1 + 6d_1 = 0 \Rightarrow 6d_0 + 6d_1 = 0 \Rightarrow d_1 = -d_0$.

Usando (4): $b_1 + c_1 + d_1 = -2$. Sustituyendo $c_1=3d_0$ y $d_1=-d_0$ y $b_1 = b_0 + 3d_0$ (de la derivada):
$b_0 + 3d_0 + 3d_0 - d_0 = -2 \Rightarrow b_0 + 5d_0 = -2$.

Con $b_0 + d_0 = 4$ restamos y obtenemos $4d_0 = -6 \Rightarrow d_0 = -\tfrac{3}{2}$.

Entonces $b_0 = 4 - d_0 = \tfrac{11}{2}$, $c_0=0$, $c_1 = 3d_0 = -\tfrac{9}{2}$, $d_1 = \tfrac{3}{2}$, $b_1 = b_0 + 3d_0 = 1$, $a_1 = 5$.

Splines finales:

$S_0(x) = 1 + \tfrac{11}{2}x - \tfrac{3}{2}x^3$ for $0\le x\le1$.

$S_1(x) = 5 + 1(x-1) - \tfrac{9}{2}(x-1)^2 + \tfrac{3}{2}(x-1)^3$ for $1\le x\le2$.

#### 2. Dados los puntos $(-1,1)$ y $(1,3)$ determine el spline cúbico sabiendo que  $f'(-1)=1$, $f'(1)=2$

Forma del tramo: $S_0(x) = a_0 + b_0(x+1) + c_0(x+1)^2 + d_0(x+1)^3$ (porque $x_0=-1$).

Condiciones:

1. Interpolación en $x=-1$: $S_0(-1)=1 \Rightarrow a_0 = 1$.

2. Interpolación en $x=1$: $S_0(1)=3 \Rightarrow 1 + 2b_0 + 4c_0 + 8d_0 = 3$.

3. Derivada: $S_0'(x)=b_0 + 2c_0(x+1) + 3d_0(x+1)^2$.

4. Condición en extremo izquierdo: $S_0'(-1)=1 \Rightarrow b_0 = 1$.

5. Condición en extremo derecho: $S_0'(1)=2 \Rightarrow 1 + 4c_0 + 12d_0 = 2$.

Sistema resultante:

De interpolación: $2 + 4c_0 + 8d_0 = 3 \Rightarrow 4c_0 + 8d_0 = 1$.

De derivada: $4c_0 + 12d_0 = 1$.

Restando: $4d_0 = 0 \Rightarrow d_0 = 0$; luego $4c_0 = 1 \Rightarrow c_0 = \tfrac{1}{4}$.

Coeficientes finales: $a_0=1,\; b_0=1,\; c_0=\tfrac{1}{4},\; d_0=0$.

Spline final: $S_0(x)=1 + 1(x+1) + \tfrac{1}{4}(x+1)^2$.

3. Diríjase al pseudocódigo del spline cúbico con frontera natural provisto en clase, en base a ese pseudocódigo complete la siguiente función:

```python
import sympy as sym
import numpy as np
from IPython.display import display

def cubic_spline(xs: list[float], ys: list[float]) -> list[sym.Symbol]:
    """
    Cubic spline interpolation ``S``. Every two points are interpolated by a cubic polynomial
    ``S_j`` of the form ``S_j(x) = a_j + b_j(x - x_j) + c_j(x - x_j)^2 + d_j(x - x_j)^3.``
    xs must be different  but not necessarily ordered nor equally spaced.
    ## Parameters
    - xs, ys: points to be interpolated
    ## Return
    - List of symbolic expressions for the cubic spline interpolation.
    """

    points = sorted(zip(xs, ys), key=lambda x: x[0])  # sort points by x

    xs = [x for x, _ in points]
    ys = [y for _, y in points]

    n = len(points) - 1  # number of splines

    if n < 2:
        # Se requiere un mínimo de 3 puntos para un spline cúbico natural no trivial.
        raise ValueError("Se requieren al menos 3 puntos para este tipo de spline.")

    h = [xs[i + 1] - xs[i] for i in range(n)]  # distances between  contiguous xs
    alpha = [0.0] * (n + 1)
    
    for i in range(1, n):
        alpha[i] = 3 / h[i] * (ys[i + 1] - ys[i]) - 3 / h[i - 1] * (ys[i] - ys[i - 1])

    l = [1.0]
    u = [0.0]
    z = [0.0]

    for i in range(1, n):
        l_i = 2 * (xs[i + 1] - xs[i - 1]) - h[i - 1] * u[i - 1]
        l.append(l_i)
        
        if l_i == 0:
             raise Exception("Sistema singular.")
             
        u.append(h[i] / l_i)
        z_i = (alpha[i] - h[i - 1] * z[i - 1]) / l_i
        z.append(z_i)

    l.append(1.0)
    z.append(0.0)
    c = [0.0] * (n + 1)

    x = sym.Symbol("x")
    splines = []
    
    for j in range(n - 1, -1, -1):
        c[j] = z[j] - u[j] * c[j + 1]
        
        b = (ys[j + 1] - ys[j]) / h[j] - h[j] * (c[j + 1] + 2 * c[j]) / 3
        d = (c[j + 1] - c[j]) / (3 * h[j])
        
        a = ys[j]
        
        print(f"S_{j}: a={a}, b={b}, c={c[j]}, d={d}")
        
        S = a + b * (x - xs[j]) + c[j] * (x - xs[j]) ** 2 + d * (x - xs[j]) ** 3

        splines.append(S)
        
    splines.reverse()
    return splines
```

4) Usando la función anterior, encuentre el spline cúbico para:
$xs = [1, 2, 3]$ y $s = [2, 3, 5]$

```python
xs = [1, 2, 3]
ys = [2, 3, 5]

splines = cubic_spline(xs=xs, ys=ys)

print("--- Expresiones Simbólicas (Forma Base) ---")
_ = [print(s) for s in splines] 

print("--- Expresiones Simbólicas (Forma Expandida) ---")
_ = [print(s.expand()) for s in splines]
```

```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt

def cubic_spline_clamped(xs: list[float], ys: list[float], fp_a: float, fp_b: float):
    """
    Cubic spline interpolation S_j(x). Aplica condiciones de frontera sujetadas (Clamped).
    fp_a = f'(x_0), fp_b = f'(x_n).
    Retorna los coeficientes (a, b, c, d) y la lista de splines simbólicos (S).
    """
    
    points = sorted(zip(xs, ys), key=lambda x: x[0])
    xs_sorted = np.array([x for x, y in points], dtype=float)
    ys_sorted = np.array([y for x, y in points], dtype=float)

    n = len(xs_sorted) - 1
    
    if n < 2:
        raise ValueError("Se requieren al menos 3 puntos.")

    h = xs_sorted[1:] - xs_sorted[:-1]
    
    alpha = np.zeros(n + 1)
    alpha[0] = 3 / h[0] * (ys_sorted[1] - ys_sorted[0]) - 3 * fp_a
    alpha[n] = 3 * fp_b - 3 / h[n-1] * (ys_sorted[n] - ys_sorted[n-1])

    for i in range(1, n):
        alpha[i] = (3/h[i] * (ys_sorted[i+1] - ys_sorted[i]) - 
                    3/h[i-1] * (ys_sorted[i] - ys_sorted[i-1])) 

    l = np.zeros(n + 1)
    u = np.zeros(n + 1)
    z = np.zeros(n + 1)
    
    l[0] = 2 * h[0]
    u[0] = 1.0
    z[0] = alpha[0] / l[0]
    
    for i in range(1, n):
        l[i] = 2 * (h[i-1] + h[i]) - h[i-1] * u[i-1]
        if l[i] == 0:
            raise Exception("Sistema no tiene solución única.")
        u[i] = h[i] / l[i]
        z[i] = (alpha[i] - h[i-1] * z[i-1]) / l[i]

    c = np.zeros(n + 1)

    l[n] = h[n-1] * (2 - u[n-1])
    c[n] = (alpha[n] - h[n-1] * z[n-1]) / l[n]
    
    for i in range(n - 1, -1, -1):
        c[i] = z[i] - u[i] * c[i+1]
        
    a = ys_sorted[:-1]
    b = np.zeros(n)
    d = np.zeros(n)
    c_spline_formula = c[:-1] 

    for i in range(n):
        d[i] = (c[i+1] - c[i]) / (3 * h[i])
        b[i] = (ys_sorted[i+1] - ys_sorted[i]) / h[i] - h[i] / 3 * (2 * c[i] + c[i+1])

    S = []
    x = sym.Symbol('x')
    for i in range(n):
        xi = xs_sorted[i]
        poly = a[i] + b[i] * (x - xi) + c_spline_formula[i] * (x - xi)**2 + d[i] * (x - xi)**3
        S.append(sym.simplify(poly))

    return a.tolist(), b.tolist(), c_spline_formula.tolist(), d.tolist(), S

def plot_spline_segments(xs, ys, splines, title):
    """Evalúa y grafica los segmentos del spline."""
    
    if not splines:
        print(f"Error: La lista de splines para '{title}' está vacía.")
        return

    x = sym.Symbol('x')
    
    plt.figure(figsize=(10, 6))
    plt.plot(xs, ys, 'o', color='blue', label='Puntos de Interpolación')
    
    for i in range(len(splines)):
        S_i = splines[i]
        S_i_func = sym.lambdify(x, S_i, 'numpy')
        x_start = xs[i]
        x_end = xs[i+1]
        x_vals = np.linspace(x_start, x_end, 100)
        y_vals = S_i_func(x_vals)
        
        label_text = 'Spline Cúbico' if i == 0 else None
        plt.plot(x_vals, y_vals, '-', color='red', label=label_text)
            
    plt.title("Spline Cúbico Clampeado")
    plt.xlabel('Eje X')
    plt.ylabel('Eje Y')
    plt.grid(True)
    plt.legend(loc='best')
    plt.show() 
```

```python
xs_1 = [1, 2, 5, 6, 7, 8, 10, 13, 17]
ys_1 = [3.0, 3.7, 3.9, 4.2, 5.7, 6.6, 7.1, 6.7, 4.5]
f_prime_a_1 = 1.0 
f_prime_b_1 = -0.67

# Llamadas a la función y graficación
_, _, _, _, splines_1 = cubic_spline_clamped(xs_1, ys_1, f_prime_a_1, f_prime_b_1)
plot_spline_segments(xs_1, ys_1, splines_1, "Spline Cúbico Clampeado: Curva 1")
```

```python
# Curva 3:
xs_3 = [27.7, 28, 29, 30]
ys_3 = [4.1, 4.3, 4.1, 3.0]
f_prime_a_3 = 0.33
f_prime_b_3 = -1.5

# Llamadas a la función y graficación
_, _, _, _, splines_3 = cubic_spline_clamped(xs_3, ys_3, f_prime_a_3, f_prime_b_3)
plot_spline_segments(xs_3, ys_3, splines_3, "Spline Cúbico Clampeado: Curva 3")
```
