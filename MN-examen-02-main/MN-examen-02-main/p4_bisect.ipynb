```python
import numpy as np
from scipy.optimize import bisect
```

```python
#Defino mi funcion
def f(x):
    return x**5 + 5.25*x**4 + 4.125*x**3 - 9.125*x**2 - 14.625*x -5.625
# Busco los intervalos donde la funcion cambia de signo
x_vals = np.linspace(-6, 1, 1000)
sing_changes = []
for i in range(len(x_vals)-1):
    if f(x_vals[i]) * f(x_vals[i+1]) < 0:
        sing_changes.append((x_vals[i], x_vals[i+1]))

# Método de bisección para encontrar las raíces
roots = []
for a, b in sing_changes:
    try:
        root = bisect(f, a, b)
        if not any(np.isclose(root, r, atol=1e-6) for r in roots):
            roots.append(root)
    except Exception:
        pass

roots = np.sort(roots)
while len(roots) < 6:
    roots = np.append(roots, np.nan)
print("Raíces encontradas:", roots)
```
