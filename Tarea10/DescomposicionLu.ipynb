### 2. Determine cuáles de las siguientes matrices son no singulares y calcule la inversa de esas matrices

![](Ejercicio2.png){width=7cm height=3cm fig-align="center"}

```python
import numpy as np

def gauss_jordan(A: np.ndarray) -> np.ndarray:
    if not isinstance(A, np.ndarray):
        A = np.array(A, dtype=float)

    n, m = A.shape
    assert m >= n, "La matriz debe tener al menos n columnas"

    for i in range(n):

        # pivoteo simple
        if A[i, i] == 0:
            for k in range(i + 1, n):
                if A[k, i] != 0:
                    A[[i, k]] = A[[k, i]]
                    break

        if A[i, i] == 0:
            raise ValueError("Matriz singular")

        # normalizar fila pivote
        A[i, :] = A[i, :] / A[i, i]

        # eliminación
        for j in range(n):
            if j != i:
                A[j, :] = A[j, :] - A[j, i] * A[i, :]

    return A
```

```python
def inversa_por_gauss_jordan(A: np.ndarray):
    det = np.linalg.det(A)
    print("Determinante =", det)

    if abs(det) < 1e-10:
        print("Matriz singular")
        return None

    n = A.shape[0]
    A_aug = np.hstack((A.astype(float), np.eye(n)))

    print("\nMatriz aumentada inicial:\n", A_aug)

    A_rref = gauss_jordan(A_aug)
    A_inv = A_rref[:, n:]

    print("\nMatriz inversa:\n", A_inv)
    return A_inv
```

```python
A = np.array([
    [4,  2,  6],
    [3,  0,  7],
    [-2, -1, -3]
], dtype=float)

print("Matriz A")
inversa_por_gauss_jordan(A)
```

```python
B = np.array([
    [1, 2,  0],
    [2, 1, -1],
    [3, 1,  1]
], dtype=float)

print("\nMatriz B")
inversa_por_gauss_jordan(B)
```

```python
C = np.array([
    [1,  1, -1,  1],
    [1,  2, -4, -2],
    [2,  1,  1,  5],
    [-1, 0, -2, -4]
], dtype=float)

print("\nMatriz C")
inversa_por_gauss_jordan(C)
```

```python
D = np.array([
    [4, 0, 0, 0],
    [6, 7, 0, 0],
    [9,11, 1, 0],
    [5, 4, 1, 1]
], dtype=float)

print("\nMatriz D")
inversa_por_gauss_jordan(D)
```

### 3. Resuelva los sistemas lineales 4 x 4 que tienen la misma matriz de coeficientes:

![](Ejercicio3.png){width=7cm height=2cm fig-align="center"}

```python
import numpy as np

def descomposicion_LU(A_in: np.ndarray):
    A = A_in.astype(float).copy() 
    n = A.shape[0]

    L = np.zeros((n, n))
    P = np.eye(n)

    for i in range(n):
        # Pivoteo parcial
        pivot = np.argmax(np.abs(A[i:, i])) + i
        
        if pivot != i:
            A[[i, pivot]] = A[[pivot, i]]
            P[[i, pivot]] = P[[pivot, i]]
            if i > 0:
                L[[i, pivot], :i] = L[[pivot, i], :i]

        L[i, i] = 1

        for j in range(i + 1, n):
            if A[i, i] == 0: 
                raise ValueError("La matriz es singular")
            
            m = A[j, i] / A[i, i]
            A[j, i:] -= m * A[i, i:]
            L[j, i] = m

    # En este punto, A se ha transformado en U
    U = A
    return P, L, U

def resolver_LU(P, L, U, b):
    n = L.shape[0]
    # 1. Ajustar b según la matriz de permutación P
    b_perm = P @ b

    # 2. Sustitución hacia adelante (Ly = Pb)
    y = np.zeros((n, 1))
    for i in range(n):
        y[i] = (b_perm[i] - L[i, :i] @ y[:i]) / L[i, i]

    # 3. Sustitución hacia atrás (Ux = y)
    x = np.zeros((n, 1))
    for i in range(n - 1, -1, -1):
        x[i] = (y[i] - U[i, i + 1:] @ x[i + 1:]) / U[i, i]

    return x
```

```python
import numpy as np

A = np.array([
    [ 1, -1,  2, -1],
    [ 1,  0, -1,  1],
    [ 2,  1,  3, -4],
    [ 0, -1,  1, -1]
], dtype=float)

P, L, U = descomposicion_LU(A)
print("Descomposición realizada con éxito.")

b1 = np.array([6, 4, -2, 5], dtype=float).reshape(-1, 1)
x1 = resolver_LU(P, L, U, b1)
print("\nSolución sistema (a):\n", x1)
```

```python
b2 = np.array([1, 1, 2, -1], dtype=float).reshape(-1, 1)

x2 = resolver_LU(P, L, U, b2)
print("Solución sistema (b):\n", x2)
```

### 4. Encuentre los valores de A que hacen que la siguiente matriz sea singular.

![](Ejercicio4.png){width=2cm height=2cm fig-align="center"}

```python
import sympy as sp

alpha = sp.symbols('alpha')
A = sp.Matrix([
    [1, -1, alpha],
    [2,  2, 1],
    [0, alpha, -sp.Rational(3, 2)]
])

A
```

```python
detA = A.det()
print(f"El determinante es: {detA}")
```

```python
sol_alpha = sp.solve(detA, alpha)

print("\nValores de alpha que hacen la matriz singular:")
for i, sol in enumerate(sol_alpha, 1):
    valor_decimal = sol.evalf()
    print(f"alpha_{i} (exacto)  : {sol}")
    print(f"alpha_{i} (decimal) : {valor_decimal:.1f}\n")
```
